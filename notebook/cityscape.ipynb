```python
# import all mandatory libraries
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
    
import torchvision
import torch 
from torchvision.datasets import Cityscapes 
import matplotlib.pyplot as plt
from torchvision.transforms import ToTensor, Compose
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
transform = Compose([
            ToTensor()  
        ])
```

```python
train_dataset = Cityscapes('../data/cityscapes', split='train', mode='fine',
                                 target_type='semantic', transform=transform, target_transform=transform)
test_dataset = Cityscapes('../data/cityscapes', split='test', mode='fine',
                                 target_type='semantic', transform=transform, target_transform=transform)
val_dataset = Cityscapes('../data/cityscapes', split='val', mode='fine',
                                 target_type='semantic', transform=transform, target_transform=transform)
```

```python
print(len(train_dataset))
print(len(test_dataset))
print(len(val_dataset))

img, _ = train_dataset[0]
print("image size = ", img.size())
```

```python
figure = plt.figure(figsize=(8, 8))
cols, rows = 3, 3
for i in range(1, cols * rows + 1):
    sample_idx = torch.randint(len(train_dataset), size=(1,)).item()
    img, label = train_dataset[sample_idx]
    figure.add_subplot(rows, cols, i)
    plt.title(label)
    plt.axis("off")
    plt.imshow(img.squeeze().numpy().swapaxis(0,1).swapaxis(1,2))
plt.show()
```
